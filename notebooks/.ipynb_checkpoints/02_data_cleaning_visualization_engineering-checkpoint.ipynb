```python
# Import pandas library
import pandas as pd
```

```python
# Define the path to the raw data file
data_path = '../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv'
```

```python
# Load Dataset into a pandas DataFrame
try:
    df = pd.read_csv(data_path)
    print("Dataset loaded successfully")
except FileNotFoundError:
    print(f"Error : Data file not found at {data_path}")
    print("Please ensure the file exists and the path is correct")
    # stop execution if file not found
    exit()
```

# Data Cleaning 

```python
# 1.Get column names and data types, non-null counts
print(" -- DataFrame Info --")
df.info()
```

```python
# 2.Handle 'TotalCharges' - convert to numeric
```

```python
# 2a.First,Identify non-numeric entries (often spaces ' ')
print("Non-Numeric entries in TotalCharges")
print(df[pd.to_numeric(df['TotalCharges'],errors='coerce').isnull()]['TotalCharges'].unique())
```

```python
# 2b.See how many such rows exist
num_non_numeric_totalcharges = pd.to_numeric(df['TotalCharges'],errors='coerce').isnull().sum()
print(f"Number of rows with non-numeric TotalCharges: {num_non_numeric_totalcharges}")
```

```python
# 2c.Examine these rows - they likely have tenure = 0
print("Rows where TotalCharges conversion failed:")
df[pd.to_numeric(df['TotalCharges'],errors='coerce').isnull()][['tenure','TotalCharges']]
```

```python
# 2d.Convert 'TotalCharges' to numeric.'coerce' will turn problematic entries into NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
# 2e.Verify the data type is now float
print(f"Data type of TotalCharges after conversion: {df['TotalCharges'].dtype}")
```

```python
# 2f.Check NaNs introduced by the conversion
print(f"NaNs in TotalCharges after conversion: {df['TotalCharges'].isnull().sum()}")
```

```python
# 2g.Impute these NaNs. Since these customers have tenure=0, their total charges logically should be 0.
# Alternatively, median/mean imputation could be used if the reason for NaNs was different.
df['TotalCharges'].fillna(0, inplace=True)
```

```python
# 2h.Verify NaNs are handled
print(f"NaNs in TotalCharges after imputation: {df['TotalCharges'].isnull().sum()}")
```

```python
# 3.Check for any other missing values
print(" -- Missing Values per column(After TotalCharges fixed) -- ")
print(df.isnull().sum())
```

```python
# 4.Drop Irrelevant columns
print(f"Dropping 'customerID'. Current columns:{df.shape[1]}")
if 'customerID' in df.columns:
    df.drop('customerID',axis=1,inplace=True)
    print(f"Dropped 'customerID'.New columns: {df.shape[1]}")
else:
    print(" 'customerID' column not found or already dropped")
```

```python
# 5.Final Verification
print("-- DataFrame Info (After Cleaning")
df.info()
```

# Feature Engineering

```python
print("Starting feature engineering...")
# Make a copy to avoid modifying the original cleaned dataframe directly if needed later
df_eng = df.copy()
```

```python
# 1. Create Tenure Groups/bins
```

```python
# 1a.Define bin edges and labels
tenure_bins = [0, 12, 24, 36, 48, 60, 72] # 0-1yr, 1-2yr, 2-3yr, etc.
tenure_labels = ['0-12 Months', '13-24 Months', '25-36 Months', '37-48 Months', '49-60 Months', '61+ Months']
```

```python
# 1b. Use pd.cut to create the bins. include_lowest=True ensures 0 is included in the first bin. right=True means bins are (edge1, edge2]
df_eng['TenureGroup'] = pd.cut(df_eng['tenure'], bins=tenure_bins, labels=tenure_labels, right=True, include_lowest=True)

print("Created 'TenureGroup' feature.")
print(df_eng['TenureGroup'].value_counts())
```

```python
# 2. Count Optional Services
```

```python
# 2a.Define the columns representing optional services
optional_service_cols = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]
```

```python
# Count how many services are 'Yes' for each customer
# (df == 'Yes') creates a boolean DataFrame, sum(axis=1) counts True values per row
df_eng['NumOptionalServices'] = df_eng[optional_service_cols].apply(lambda row: sum(row == 'Yes'), axis=1)

print("\nCreated 'NumOptionalServices' feature.")
print(df_eng['NumOptionalServices'].value_counts())
```

```python
# 3. (Optional Exploration) Monthly Charges per Tenure
# Avoid division by zero for tenure=0 customers (though TotalCharges fix might have handled this implicitly if tenure 0 -> totalcharges 0)
# Let's use a small epsilon to avoid division by zero, although tenure 0 cases should have TotalCharges 0 anyway.
# df_eng['MonthlyChargesPerTenure'] = df_eng['TotalCharges'] / (df_eng['tenure'] + 0.001)
# -- Decided to skip this for now as it might be too correlated with existing features --
```

```python
# 4.Flag for No Internet Add-ons (for customers WITH Internet Service)
```

```python
# 4a.First identify customers with Internet Service
has_internet = df_eng['InternetService'] != 'No'
```

```python
# 4b.Then check if they have NONE of the specific add-ons
no_addons = df_eng['NumOptionalServices'] == 0
df_eng['HasInternet_NoAddons'] = (has_internet & no_addons)

print("\nCreated 'HasInternet_NoAddons' feature (True/False).")
print(df_eng['HasInternet_NoAddons'].value_counts())


print("\nFeature engineering complete. Displaying first few rows with new features:")
display(df_eng[['tenure', 'TenureGroup', 'NumOptionalServices', 'InternetService', 'HasInternet_NoAddons', 'Churn']].head())
```

```python
# 4c.Check data types of new columns
print("\nInfo after adding features:")
df_eng.info()
```

```python
# 1.TenureGroup vs Churn Rate
plt.figure(figsize=(10, 5))
# Calculate churn rate per group
churn_rate_tenure_group = df_eng.groupby('TenureGroup')[target].value_counts(normalize=True).unstack()
if 'Yes' not in churn_rate_tenure_group.columns: churn_rate_tenure_group['Yes'] = 0
churn_rate_tenure_group = churn_rate_tenure_group.reset_index()
# Plot
sns.barplot(data=churn_rate_tenure_group, x='TenureGroup', y='Yes')
plt.title('Churn Rate by Tenure Group')
plt.ylabel('Churn Rate')
plt.xlabel('Tenure Group')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2. NumOptionalServices vs Churn Rate
plt.figure(figsize=(10, 5))
# Calculate churn rate per number of services
churn_rate_num_services = df_eng.groupby('NumOptionalServices')[target].value_counts(normalize=True).unstack()
if 'Yes' not in churn_rate_num_services.columns: churn_rate_num_services['Yes'] = 0
churn_rate_num_services = churn_rate_num_services.reset_index()
# Plot
sns.barplot(data=churn_rate_num_services, x='NumOptionalServices', y='Yes')
plt.title('Churn Rate by Number of Optional Services')
plt.ylabel('Churn Rate')
plt.xlabel('Number of Optional Services')
plt.tight_layout()
plt.show()
```

```python
# 3. HasInternet_NoAddons vs Churn Rate
plt.figure(figsize=(6, 4))
# Calculate churn rate
churn_rate_no_addons = df_eng.groupby('HasInternet_NoAddons')[target].value_counts(normalize=True).unstack()
if 'Yes' not in churn_rate_no_addons.columns: churn_rate_no_addons['Yes'] = 0
churn_rate_no_addons = churn_rate_no_addons.reset_index()
# Plot
sns.barplot(data=churn_rate_no_addons, x='HasInternet_NoAddons', y='Yes')
plt.title('Churn Rate for Customers with Internet but No Addons')
plt.ylabel('Churn Rate')
plt.xlabel('Has Internet Service but No Addons')
plt.xticks([False, True], ['False', 'True']) # Ensure labels are clear bools
plt.tight_layout()
plt.show()
```
