```python
# Import pandas library
import pandas as pd
```

```python
# Define the path to the raw data file
data_path = '../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv'
```

```python
# Load Dataset into a pandas DataFrame
try:
    df = pd.read_csv(data_path)
    print("Dataset loaded successfully")
except FileNotFoundError:
    print(f"Error : Data file not found at {data_path}")
    print("Please ensure the file exists and the path is correct")
    # stop execution if file not found
    exit()
```

```python
# 1.Display the first 5 rows
print("\n-- First 5 Rows --")
display(df.head())
```

```python
# 2.Display the last 5 rows
print("\n-- Last 5 Rows--")
display(df.tail())
```

```python
# 3. Get DataFrame shape (rows,columns)
print("\n -- DataFrame Shape --")
print(f"Rows:{df.shape[0]},Columns:{df.shape[1]}")
```

```python
# 4.Get column names and data types, non-null counts
print(" -- DataFrame Info --")
df.info()
```

```python
# 5.Get summary statistics for numerical features
print("-- Numerical Features Summary --")
display(df.describe())
```

```python
# 6.Get summary statistics for categorical features
print("-- Categorical Features Summary --")
# Include 'object' columns and potentially boolean columns if any
display(df.describe(include=['object','bool']))
```

```python
# 7. List all column names
print("-- Column Names --")
print(df.columns.tolist())
```

```python
# 8. Check for missing values (Sum per column)
print("-- Missing Values per Column --")
print(df.isnull().sum())
```

```python
# 9.Check the distribution of the target variable 'Churn'
print("-- Churn Distribution --")
print(df['Churn'].value_counts())
print("-- Churn Distribution (%) --")
print(df['Churn'].value_counts(normalize=True)*100)
```
